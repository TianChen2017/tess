```python
import tess
```

```python
nm = 8
ptsr = [tuple(array((l,m,n))+.5) + ((l+m+n) % 2,)
       for l in range(nm) for m in range(nm) for n in range(nm)]
ptsr = np.array(ptsr)
pts = ptsr[:, :3]
r = [.5 if r < 1 else .6 for r in ptsr[:, 3]]
```

```python
cells1 = tess.Container(pts, nm, periodic=False)
c = cells1[0]
len(cells1), str(c), repr(c)
```

```python
cells2 = tess.Container(pts, nm, radii=r, periodic=True)
len(cells2)
```

### Neighbors
We have calculated all the cells and neighbors, so now lets go through and make some graphs

First, how many neighbors does each cell have?

```python
for cells in (cells1, cells2):
    alln = []
    for v in cells:
        alln.append(len(v.neighbors()))

    print(ufloatms(alln))
    for n in sorted(set(alln)):
        print('{:3d}:{:3d}'.format(n, alln.count(n)))
```

How much volume does each cell have?

```python
for cells in (cells1, cells2):
    allVs = []
    for v in cells:
        allVs.append(v.volume())

    print(ufloatms(allVs))
    for n in sorted(set(['{:8.3g}'.format(V) for V in allVs])):
        print('{:8s}:{:3f}'.format(n, allVs.count(n)))
```

Plot the connections from a large cell (left) and from a small cell (right)

```python
from mpl_toolkits.mplot3d import Axes3D

for cells in (cells1, cells2):
    (fig, (ax1, ax2)) = plt.subplots(1,2,figsize=(10,4), subplot_kw=dict(projection='3d'))

    p1, p2 = array((1.5,1.5,1.5)), array((1.5,1.5,2.5))
    for v in cells:
        if np.allclose(p1, v.pos()):
            ax = ax1
        elif np.allclose(p2, v.pos()):
            ax = ax2
        else:
            continue

        x1,y1,z1 = v.pos()
        xyz2 = [cells[m].pos() for m in v.neighbors()]

        for x2,y2,z2 in sorted(xyz2):
            ls = 'k-'
            if z1 - z2 > z1*1e-3: ls = 'r-'
            elif z1 - z2 < -z1*1e-3: ls = 'b-'
            ax.plot([x1,x2], [y1,y2], [z1,z2], ls)
```

### FCC

```python
FCC_r1 = np.array((1,0,0))
FCC_r2 = np.array((.5,sqrt(3.)/2., 0))
FCC_r3 = np.array((.5,sqrt(3.)/6., sqrt(6.)/3))

FCC_x0 = np.array((0,0,0))
FCC_x1 = np.array((0,0.5,0.5))
FCC_x2 = np.array((0.5,0,0.5))
FCC_x3 = np.array((0.5,0.5,0))
```

```python
[np.sum(r**2) for r in (FCC_r1,FCC_r2,FCC_r3)]
```

```python
def not_fcc(n):
    x0 = np.array((0.5,0.5,0.5))
    r1 = np.array((1,0,0))
    r2 = np.array((.5,sqrt(3.)/2., 0))
    r3 = np.array((.5,sqrt(3.)/6., sqrt(6.)/3))
    return np.array([x0 + (i*r1) + (j*r2) + (k*r3) for k in range(n) for j in range(n) for i in range(n)])

def fcc(n):
    dxs = np.array([(i,j,k) for k in range(n) for j in range(n) for i in range(n)])
    xs = np.array([[x+dx for x in (FCC_x0, FCC_x1, FCC_x2, FCC_x3)] for dx in dxs])
    return np.concatenate(xs)
```

```python
def hcp(n):
    assert n % 2 == 0
    arr = np.array([(
                        2*i + (j%2) + (k%2),
                        sqrt(3.0)*(j + (k%2)/3.),
                        sqrt(24.)*k/3)
                     for k in range(n) for j in range(n) for i in range(n)])
    return arr/2.
```

```python
n = 4
xyz = hcp(n)
xyz.shape
```

```python
#limits = np.array((n, n*sqrt(3.)/2., n*sqrt(6.)/3)) # not_fcc
limits = np.array((n,n,n)) # fcc
limits = np.array((n, n*sqrt(3.0)/2, n*sqrt(6.)/3.))
cells = tess.Container(xyz, limits=limits, periodic=True)
```

```python
allns, allVs = [], []
for v in cells:
    allns.append(len(v.neighbors()))
    allVs.append(v.volume())

print(ufloatms(allVs))
vlst = ['{:8.3g}'.format(V) for V in allVs]
for n in sorted(set(vlst)):
    print('{:8s}:{:3d}'.format(n, vlst.count(n)))

print(ufloatms(allns))
for n in sorted(set(allns)):
    print('{:3d}:{:3d}'.format(n, allns.count(n)))
```

```python
def diff(a,b, limits=limits):
    return np.remainder(limits/2. + a - b, limits) - limits/2.
```
