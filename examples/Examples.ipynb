```python
import pyvoro
```

```python
nm = 8
fnm = float(nm)
c = pyvoro.ContainerPoly(0.,fnm,0.,fnm,0.,fnm, # limits of the box
                     2,2,2, # grid size
                     True,True,True, # periodic in all directions
                     20 # room for 20 points
                     )
```

```python
# Make points in a grid, with every other point having a larger radius
pts = [(tuple((array((l,m,n))+.5)),(l+m+n) % 2) 
       for l in range(nm) for m in range(nm) for n in range(nm)]
ptsr = [(x,y,z, 1.0 if r < 1 else 1.2) for ((x,y,z),r) in pts]
# A list of (x,y,z,r): position and radius of each point
c.add(ptsr)
# Get the voronoi cells
cells = c.get_cells()
cellsbyn = {v.id: v for v in cells}
```

### Neighbors
We have calculated all the cells and neighbors, so now lets go through and make some graphs

First, how many neighbors does each cell have?

```python
alln = []
for v in cells:
    alln.append(len(v.neighbors()))

print(ufloatms(alln))
for n in sorted(set(alln)):
    print('{:3d}:{:3d}'.format(n, alln.count(n)))
```

How much volume does each cell have?

```python
allVs = []
for v in cells:
    allVs.append(v.volume())

print(ufloatms(allVs))
for n in sorted(set(['{:8.3g}'.format(V) for V in allVs])):
    print('{:8s}:{:3f}'.format(n, allVs.count(n)))
```
